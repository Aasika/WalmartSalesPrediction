6. Handling Outliers

```python
print(data['fare'].mean())
plt.hist(data['fare'])
#Concluding that, fare greater than 150 are outliers
```

```python
#Use of zscore to eliminate outliers
z_scores = (data['fare'] - data['fare'].mean()) / data['fare'].std()
threshold = 2 #95%
outlier_mask = (z_scores > threshold) | (z_scores < -threshold)

# Remove outliers from the DataFrame
data = data[~outlier_mask]
print(data.shape)
print(data['fare'].mean())
plt.hist(data['fare'])
#Outliers like 150+ are removed.
```

```python
#Similar action for age
print(data['age'].mean())
plt.hist(data['age'])
```

```python
#Use of zscore to eliminate outliers
z_scores = (data['age'] - data['age'].mean()) / data['age'].std()
threshold = 3 #99%
outlier_mask = (z_scores > threshold) | (z_scores < -threshold)

# Remove outliers from the DataFrame
data = data[~outlier_mask]
print(data.shape)
print(data['age'].mean())
plt.hist(data['age'])
#Outliers like 70+ are removed.
```
